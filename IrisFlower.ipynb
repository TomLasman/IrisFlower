```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['axes.titlesize']= 28
plt.rcParams['axes.labelsize']= 16

from sklearn import datasets
#Principal Component Analysis
from sklearn.decomposition import PCA 
#Linear Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
iris = datasets.load_iris()

#4 attributes: sepal length, sepal width, petal length and petal width.
data = iris.data
#3 kind of Iris flowers (Setosa = 0, Versicolourand = 1 Virginica = 2) 
target = iris.target
#names of Iris flowers
target_names = iris.target_names

colors = ['navy', 'green', 'darkorange']
kindOfIris = [0, 1, 2]

figure, ((a1, a2)) = plt.subplots(1,2) # figure get the resolution 

for color, i, target_name in zip(colors, kindOfIris, target_names):
     a1.scatter(data[target == i, 0], data[target == i, 1], color=color, label=target_name)
     
a1.set(xlabel='[cm]', ylabel='[cm]',title='IRIS Sepal Dataset')
a1.legend(loc='best', shadow=False, scatterpoints=1)


for color, i, target_name in zip(colors, kindOfIris, target_names):
     a2.scatter(data[target == i, 2], data[target == i, 3], color=color, label=target_name)

a2.legend(loc='best', shadow=False, scatterpoints=1)
a2.set(xlabel='[cm]', ylabel='[cm]',title="IRIS petal Dataset")

figure.tight_layout()


```

```python
pca = PCA(n_components=2) 
X_r = pca.fit(data).transform(data)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(data, target).transform(data)
```

```python
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
#LineWidth
lw = 2 
kindOfIris = [0, 1, 2]

for color, i, target_name in zip(colors,kindOfIris , target_names):
    plt.scatter(X_r[target == i, 0], X_r[target == i, 1], color=color, alpha=.8, lw=lw, label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS Dataset')
```

```python
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[target == i, 0], X_r2[target == i, 1], alpha=.8, color=color, label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')
```
